```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random as rd

np.random.seed(42)
#data = pd.read_csv(r"C:\Users\canar\OneDrive - Universidad de Nariño\UNIVERSIDAD\SEMESTRE X\INTELIGENCIA ARTIFICIAL - ELECTIVA\CORTE I\data.csv")
datos = pd.read_csv('data.csv')
data = np.array(datos)


#w1, w2, b_test = 0.6,1,12

X = data[:,[0,1]]
y = data[:,2]


x_max = max(X.T[0])
W_test = np.array(np.random.rand(2,1))
b_test = np.random.rand(1)[0] + x_max
```

```python
W = [0.5,0.5]
#funcion sigmoide 
def sigmoide(score):
    return np.exp(score)/(1+np.exp(score))

#funcion softmax
def softmax(z):
    exp_z = np.exp(z)
    return exp_z / np.sum(exp_z)

#Funcion  escalon
def escalon(t):
    if t>=0:
        return 1
    return 0

#funcion para prediccion
def prediccion(X,W,b):
    #print('score',(np.matmul(X,W)+b)[0])
    return (np.matmul(X,W)+b)[0]

```

```python
#Agrupar los valores de la predicción ( WX +b) en un vector 

p_l = []
for i in range(len(X)):
    p=prediccion(X[i,[0,1]],W_test,b_test)
    p_l.append(p)

 




```

```python
#Codigo de grafico fucion sigmoide 
ss = sigmoide(p_l)

x111=np.linspace(min(ss),max(ss))                              #vector entre 0 10 con 10 espacios entre si 
y111=np.linspace(min(ss),max(ss))
A,B=np.meshgrid(x111,y111)                             # malla necesaria para el contorno 


z=np.array((W[0]*A)+(W[1]*B)-0.5)                    # funcion lineal con base en la malla creada 
#print(z)

plt.figure(figsize=(9,6))
#plt.plot(c1['x1'],c1['x2'],'o',label="1")
#plt.plot(c2['x1'],c2['x2'],'+',label="0")
plt.legend()
plt.contourf(A, B, z)                                  # la funcion contourf permite generar mapas de contorno con las coordenadas que se le proporcionan 
plt.xlim([min(ss),max(ss)])
plt.ylim([min(ss),max(ss)])

ss = sigmoide(p_l)
#Graficar los datos, separando las clases 
plt.scatter(ss,ss,color = "r")
plt.xlim([min(ss),max(ss)])
plt.ylim([min(ss),max(ss)])
plt.title("Distribución de probabilidad función sigmoide")
```

```python
#Codigo de grafico fucion softmax
sf = softmax(p_l)
x111=np.linspace(min(sf),max(sf))                              #vector entre 0 10 con 10 espacios entre si 
y111=np.linspace(min(sf),max(sf))
A,B=np.meshgrid(x111,y111)                             # malla necesaria para el contorno 


z=np.array((W[0]*A)+(W[1]*B)-0.5)                    # funcion lineal con base en la malla creada 
#print(z)

plt.figure(figsize=(9,6))
#plt.plot(c1['x1'],c1['x2'],'o',label="1")
#plt.plot(c2['x1'],c2['x2'],'+',label="0")
plt.legend()
plt.contourf(A, B, z)                                  # la funcion contourf permite generar mapas de contorno con las coordenadas que se le proporcionan 
plt.xlim([min(sf),max(sf)])
plt.ylim([min(sf),max(sf)])

#Graficar los datos, separando las clases 

plt.scatter(sf,sf,color = "r")
plt.xlim([min(sf),max(sf)])
plt.ylim([min(sf),max(sf)])
plt.title("Distribución de probabilidad función softmax")
```
